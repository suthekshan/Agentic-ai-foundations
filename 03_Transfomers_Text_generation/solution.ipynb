```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch

# Task 2: Load the pre-trained GPT-2 model and tokenizer
# ANSWER: The standard identifier is "gpt2"
model_name = "gpt2"

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

model.eval() 
print("Model Loaded Successfully!")
```

```python
# Task 3: Convert text into numbers (vectors)
prompt = "Artificial intelligence will"

# ANSWER: Pass the 'prompt' variable into the tokenizer
inputs = tokenizer(prompt, return_tensors="pt")

print("Token IDs:", inputs['input_ids'])
# Expected Output: Tensor with integers like [50256, 38112, ...]
```

```python
# Task 4: Generate text!
outputs = model.generate(
    **inputs, 
    # ANSWER: max_length must be greater than the input length (e.g., 50)
    max_length=50,     
    do_sample=True,     
    temperature=0.7     
)

generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Text:\n" + 100*"-" + "\n")
print(generated_text)
```

```python
# SCENARIO A: The Boring Fact-Checker
prompt_fact = "The Earth revolves around the"
input_fact = tokenizer(prompt_fact, return_tensors="pt")

# ANSWER KEY: Low Temperature + Low Top_K = Deterministic/Factual
outputs = model.generate(
    **input_fact, 
    max_length=20,
    do_sample=True,
    
    # Low temp makes the probability distribution sharp (picks only high prob words)
    temperature=0.1,   
    
    # Low k restricts the model to only the top 10 most likely words
    top_k=10          
)

print("Scenario A Result:", tokenizer.decode(outputs[0], skip_special_tokens=True))
# Likely Result: "The Earth revolves around the sun..."
```

```python
# SCENARIO B: The Creative Storyteller
prompt_story = "Once upon a time in a galaxy far away,"
input_story = tokenizer(prompt_story, return_tensors="pt")

# ANSWER KEY: High Temperature + Higher Top_K = Creative/Diverse
outputs = model.generate(
    **input_story, 
    max_length=100,
    do_sample=True,
    
    # High temp flattens the distribution, giving lower prob words a chance
    temperature=0.9,   
    
    # Higher k allows more vocabulary choices
    top_k=50           
)

print("Scenario B Result:", tokenizer.decode(outputs[0], skip_special_tokens=True))
# Likely Result: A random, interesting sci-fi story.
```

```python
# SCENARIO C: Fixing the "Broken" Bot
prompt_debug = "Machine learning is"
input_debug = tokenizer(prompt_debug, return_tensors="pt")

# ANSWER KEY: Restore balanced default values
outputs = model.generate(
    **input_debug, 
    max_length=50,      
    do_sample=True,
    
    # Fixed: 0.7 is a good balance. 0.01 was too repetitive.
    temperature=0.7,    
    
    # Fixed: 50 is standard. 1 was too restrictive (only picked 1 word).
    top_k=50            
)

print("Fixed Result:", tokenizer.decode(outputs[0], skip_special_tokens=True))
```
